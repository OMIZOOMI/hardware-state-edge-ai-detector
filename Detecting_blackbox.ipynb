```python
import json
import os
from roboflow import Roboflow
from ultralytics import YOLO


# Secureling load API keys from JSON

print("Loading credentials from JSON...")
try:
    with open('credentials.json', 'r') as file:
        creds = json.load(file)
        API_KEY = creds['roboflow_api_key']
        WORKSPACE = creds['workspace_name']
        PROJECT = creds['project_name']
        VERSION = creds['version_number']
        print("Credentials loaded successfully!")
except FileNotFoundError:
    print("Error: credentials.json file not found.")
    exit()


# Downloading Dataset from Roboflow

print("Downloading dataset from Roboflow...")
rf = Roboflow(api_key="lSuIA01kEFWwE7CiSygm")
project = rf.workspace("oms-workspace-gpkjf").project("modem-led-detection")
dataset = project.version(1).download("yolov8")


# Training YOLOv8 Nano Model

print("Initializing YOLOv8 Nano model...")
model = YOLO('yolov8n.pt')

print("Starting training phase...")

results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=25,
    imgsz=640,
    plots=True
)


# Exporting to TFLite (The 'ftlight' requirement)

print("Exporting model to TensorFlow Lite (TFLite)...")
exported_model_path = model.export(format="tflite")

print(f"Success! Your TFLite model is saved at: {exported_model_path}")
print("You are ready to deploy to the edge!")
```
